## Different kinds of plots

### <font color=blue> Line plots

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'year': [2010, 2011, 2012, 2013, 2014, 2015],
        'sales': [100, 150, 120, 180, 200, 250]}
df = pd.DataFrame(data)

# Create a line plot
plt.plot(df['year'], df['sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over Time')

# Display the plot
plt.show()
```

```python
# multiple lines
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'year': [2010, 2011, 2012, 2013, 2014, 2015],
        'sales1': [100, 150, 120, 180, 200, 250],
        'sales2': [80, 120, 110, 160, 190, 220]}
df = pd.DataFrame(data)

# Create a line plot with multiple lines
plt.plot(df['year'], df['sales1'], label='Sales 1')
plt.plot(df['year'], df['sales2'], label='Sales 2')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over Time')
plt.legend()

# Display the plot
plt.show()
```

### <font color=blue> Bar plots

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'year': [2010, 2011, 2012, 2013, 2014, 2015],
        'sales': [100, 150, 120, 180, 200, 250]}
df = pd.DataFrame(data)

# Create a bar plot
plt.bar(df['year'], df['sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales by Year')

# Display the plot
plt.show()
```

### <font color=blue> Scatter plots

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'x': [1, 2, 3, 4, 5],
        'y': [2, 4, 6, 8, 10]}
df = pd.DataFrame(data)

# Create a scatter plot
plt.scatter(df['x'], df['y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')

# Display the plot
plt.show()
```

```python
# with trend line
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a DataFrame
data = {'x': [1, 2, 3, 4, 5],
        'y': [2, 4, 6, 8, 10]}
df = pd.DataFrame(data)

# Create a scatter plot with trend line
sns.regplot(x='x', y='y', data=df)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Trend Line')

# Display the plot
plt.show()
```

### <font color=blue> Histogram

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# Create a DataFrame
df = pd.DataFrame({'values': data})

# Create a histogram
plt.hist(df['values'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')

# Display the plot
plt.show()
```

### <font color=blue> Box plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a DataFrame
data = {'category': ['A', 'A', 'A', 'B', 'B', 'B', 'B'],
        'value': [10, 15, 20, 25, 30, 35, 40]}
df = pd.DataFrame(data)

# Create a box plot
sns.boxplot(x='category', y='value', data=df)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Box Plot')

# Display the plot
plt.show()
```

### <font color=blue> Heatmaps

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a DataFrame
data = {'A': [1, 2, 3],
        'B': [4, 5, 6],
        'C': [7, 8, 9]}
df = pd.DataFrame(data)

# Create a heatmap
sns.heatmap(df, annot=True, cmap='YlGnBu')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.title('Heatmap')

# Display the plot
plt.show()
```
